```python
dataset = pp.load_dataset("../adl-piano-midi")
JazzSongs = pp.files_to_songs(dataset["Jazz"])
PopSongs = pp.files_to_songs(dataset["Pop"])

channel_to_ind, ind_to_channel, note_to_ind, ind_to_note, velocity_to_ind, ind_to_velocity = pp.dicts_from_songs(JazzSongs | PopSongs)

time_range = pp.ranges_from_songs(JazzSongs | PopSongs)

n_Channels = len(channel_to_ind)
n_Notes = len(note_to_ind)
n_Velocities = len(velocity_to_ind)

print("\nNumber of channels   :",n_Channels,"\nNumber of notes      :",n_Notes,"\nNumber of velocities :",n_Velocities,"\nTime range           :",time_range[0],time_range[1])
```

```python
def plot_comp(dataname,losses,channels_acc,notes_acc,velocities_acc,times_acc):
    x = np.arange(5)
    xtick = ["Jazz only","Pop only","Jazz&Pop","Pop&Jazz","Mixed"]
    plt.figure()
    plt.bar(x,losses,tick_label=xtick)
    plt.xlabel("Model")
    plt.ylabel("Loss")
    plt.title("Performance of the different models on the "+dataname+" validation set")
    plt.show()
    
    fig, axs = plt.subplots(2, 2, figsize=(9, 8))

    axs[0,0].bar(x, channels_acc, tick_label=xtick)
    axs[0,0].set_title("Channels")
    axs[0,0].set_xlabel("Model")
    axs[0,0].set_ylabel("Accuracy")

    axs[0,1].bar(x, notes_acc, tick_label=xtick)
    axs[0,1].set_title("Notes")
    axs[0,1].set_xlabel("Model")
    axs[0,1].set_ylabel("Accuracy")

    axs[1,0].bar(x, velocities_acc, tick_label=xtick)
    axs[1,0].set_title("Velocity")
    axs[1,0].set_xlabel("Model")
    axs[1,0].set_ylabel("Accuracy")

    axs[1,1].bar(x, times_acc, tick_label=xtick)
    axs[1,1].set_title("Time")
    axs[1,1].set_xlabel("Model")
    axs[1,1].set_ylabel("Accuracy")

    fig.suptitle("Performance of the different models on the "+dataname+" validation set")
    fig.tight_layout(rect=[0, 0, 1, 1])
    plt.show()

%matplotlib qt
plot_comp("Jazz",losses_J,channels_acc_J,notes_acc_J,velocities_acc_J,times_acc_J)
plot_comp("Pop",losses_P,channels_acc_P,notes_acc_P,velocities_acc_P,times_acc_P)
plot_comp("Mixed",losses_M,channels_acc_M,notes_acc_M,velocities_acc_M,times_acc_M)
```
