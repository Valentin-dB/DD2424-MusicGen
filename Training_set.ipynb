```python
def keep_note_on(file_path):
    midi = mido.MidiFile(file_path)
    notes = []
    for message in midi:
        if message.type == 'note_on':
            notes.append(message)
    return notes
```

```python
def midi_to_notes(midi_path):
    midi = mido.MidiFile(midi_path)
    notes = []
    for msg in midi:
        if msg.type == 'note_on':
            notes.append(msg.note)
    return notes
```

```python
def times_between_notes(midi_path):
    midi = mido.MidiFile(midi_path)
    times = []
    for msg in midi:
        if msg.type == 'note_on':
            times.append(msg.time)  
    return times
```

```python
def create_dict(notes):
    dict_notes = {}
    for note in notes:
        if note not in dict_notes:
            dict_notes[note] = len(dict_notes)
    return dict_notes

dict_notes = create_dict(notes)
dict_notes
```

```python
def one_hot_encoding(dict_notes,notes):
    k = len(dict_notes)
    one_hot = np.zeros((len(notes), k))
    for i in range(len(notes)):
        one_hot[i][dict_notes[notes[i]]] = 1
    return one_hot

one_hot = one_hot_encoding(dict_notes,notes)
one_hot
```
